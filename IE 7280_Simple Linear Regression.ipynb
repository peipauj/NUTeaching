## Correlation Analysis

This section shows the correlation analysis by using biking example. In particular, we compare the correlation coefficients calculated directly from the formula vs R output.

```r
# ======= Example: Maintenance Costs ================
machine <- data.frame(age=c(2,5,9,3,8), cost=c(6,13,23,5,22))

plot(machine)

# calculate the correlation:
x = machine$age
y = machine$cost
x_bar = mean(x)
y_bar = mean(y)
r = sum((x-x_bar)*(y-y_bar))/sqrt(sum((x-x_bar)^2)*sum((y-y_bar)^2))

cor(machine)
```

## Building simple linear regression model in R

The section now demonstrates how to build simple linear regression model using a click ball point pens example. Togethe with regression model, the residual and correlation analysis is also shown below.

```r

# ======= Example: Click Ball Point Pens ===========
click = read.table("/Users/ppei/NE Teaching/Fall 2019/IE 7280 Statistical Models/Lecture Notes/Wei version/A01_Simple Linear Regression/click.txt", header=T, sep="\t")
n = nrow(click)
summary(click)

plot(click[, 1:4])  # plot scatter matrix for the data
# calculate the sample correlation matrix
round(cor(click), 4)

# plot the scatter plot of the data
plot(click$ad,click$sales)

# fit the linear model
fit = lm(sales~ad, click)
plot(click$ad, click$sales,xlab="# TV Spots", ylab="Sales")
abline(fit)

# summary the fitted linear model
summary(fit)

confint(fit)

2*(1-pt(11.43,n-2))

predict(fit, data.frame(ad=5))

predict(fit, data.frame(ad=5), interval="prediction")
predict(fit, data.frame(ad=5), interval="confidence")


# fitted values and residuals at x1,x2,...,xn:
fit$fit      # \widehat{y}_i
fit$res      # \widehat{e}_i=y_i-\widehat{y}_i
plot(fit$fit,fit$res,ylab="Residuals",xlab="Fitted response")
plot(fit$fit,abs(fit$res),ylab="Abs Residuals",xlab="Fitted response")
plot(fit$fit,(fit$res)^2,ylab="Abs Residuals",xlab="Fitted response")

qqnorm(fit$res)
qqline(fit$res)

summary(lm(abs(fit$res)~fit$fit))

fit2 = lm(log10(sales)~ad,click)
summary(lm(abs(fit2$res)~fit2$fit))

# semistudentized residuals
stdRes = fit$res/59.56
plot(fit$fit,stdRes)

# study residual for diagnotics 
plot(fit)
```

## ANOVA
The example below shows the Analysis of Variances (ANOVA) including SSE, SST and SSR.

```r
# ====== Example: Maintenance costs for ANOVA =======
machine = data.frame(age=c(2,5,9,3,8), cost=c(6,13,23,5,22))
fit = lm(cost~age, data=machine)
summary(fit)

plot(machine)
abline(fit)

# caculate SSE,SST,SSR
sum(fit$residuals^2) # find SSE
sum((machine$cost-mean(machine$cost))^2) # find SST
sum((fit$fitted.values-mean(machine$cost))^2) # find SSR

# obtain ANOVA table:
anova(fit)

plot(machine)
abline(fit)
points(machine$age, fit$fitted.values, pch=2)
ybar = mean(machine$cost)
abline(h=ybar)

segments(machine$age, machine$cost, machine$age, fit$fitted.values, lwd=4) #SSE
segments(machine$age-.02, machine$cost, machine$age-.02, ybar, col=2, lty=2, lwd=4) #SST
#SSR
segments(machine$age+.02, ybar, machine$age+.02, fit$fitted.values, col=3, lty=3, lwd=4)
legend(2, 22, c("SSE", "SST", "SSR"), col=1:3, lty=1:3)
```
